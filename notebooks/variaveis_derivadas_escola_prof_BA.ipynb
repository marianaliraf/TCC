```python
def variaveis_pos(dataset, nome_coluna, nome_nova_coluna):
    variavel_derivada = dataset[['CO_ENTIDADE', nome_coluna]].loc[professores[nome_coluna] == 1]

    variavel_derivada = variavel_derivada.groupby(by='CO_ENTIDADE').count()

    variavel_derivada.rename(columns={nome_coluna:nome_nova_coluna}, inplace=True)

    return variavel_derivada
```

```python
#Criando variaveis derivadas sobre especializacao

colunas = {'IN_ESPECIALIZACAO':'QtdeDocentesEsp', 
           'IN_MESTRADO': 'QtdeDocentesMest', 
           'IN_DOUTORADO': 'QtdeDocentesDout',
           'IN_POS_NENHUM': 'QtdeDocentesSemPos',
           'IN_ESPECIFICO_ANOS_FINAIS': 'QtdeDocentesFCEM',
           'IN_DISC_DIVER_SOCIO_CULTURAL': 'QtdeDocentesDivSocCulta',
           'IN_DISC_ATENDIMENTO_ESPECIAIS' : 'QtdeDocentesAtendEsp'
           }

   # 'TP_TIPO_DOCENTE',
   # 'TP_TIPO_CONTRATACAO' 
   # 'CO_PESSOA_FISICA': 'QtdeDocentesTotal',
   # 'ID_TURMA': 'QtdeTurmasProf'  ,

pos_graduacao = pd.DataFrame()

for coluna, nova_coluna in colunas.items():
    if (pos_graduacao.empty):
        pos_graduacao = variaveis_pos(professores, coluna, nova_coluna)
    else:
        pos_graduacao = pos_graduacao.join(variaveis_pos(professores, coluna, nova_coluna))
    

pos_graduacao = pos_graduacao.fillna(0).astype(int)
```

####Criando variaveis derivadas quantidade de docentes, turmas, disciplinas voltadas a diversidade e atendimento especial


```python
sdf = psql.sqldf("select CO_ENTIDADE, count(CO_PESSOA_FISICA) as QtdeDocentesTotal, count(ID_TURMA) as QtdeTurmasProf from professores group by CO_ENTIDADE")

sdf.set_index('CO_ENTIDADE', inplace=True)
```

```python
variaveis_docentes = pd.DataFrame()

variaveis_docentes = pos_graduacao.join(sdf)
```

####Criando variaveis com 'TP_TIPO_DOCENTE' 

```python


docentes = psql.sqldf("select CO_ENTIDADE, count(TP_TIPO_DOCENTE) as QtdeDocentes from professores where TP_TIPO_DOCENTE = 1 group by CO_ENTIDADE")

docentesAuxAssist = psql.sqldf("select CO_ENTIDADE, count(TP_TIPO_DOCENTE) as QtdeDocentesAuxAssist from professores where TP_TIPO_DOCENTE = 2 group by CO_ENTIDADE")

docentesMonit = psql.sqldf("select CO_ENTIDADE, count(TP_TIPO_DOCENTE) as QtdeDocentesMonit from professores where TP_TIPO_DOCENTE = 3 group by CO_ENTIDADE")

docentes.set_index('CO_ENTIDADE', inplace=True)

docentesAuxAssist.set_index('CO_ENTIDADE', inplace=True)

docentesMonit.set_index('CO_ENTIDADE', inplace=True)
```

```python
variaveis_docentes = variaveis_docentes.join(docentes)
variaveis_docentes = variaveis_docentes.join(docentesAuxAssist)
variaveis_docentes  = variaveis_docentes.join(docentesMonit)

variaveis_docentes
```

#### Criando variaveis derivadas de'TP_TIPO_CONTRATACAO' 

```python
docentesEfetivos = psql.sqldf("select CO_ENTIDADE, count(TP_TIPO_CONTRATACAO) asQtdeDocentesEfetivos from professores where TP_TIPO_CONTRATACAO = 1 group by CO_ENTIDADE")

docentesTempo = psql.sqldf("select CO_ENTIDADE, count(TP_TIPO_CONTRATACAO) as QtdeDocentesTempo from professores where TP_TIPO_CONTRATACAO = 2 group by CO_ENTIDADE")

docentesTerceir = psql.sqldf("select CO_ENTIDADE, count(TP_TIPO_CONTRATACAO) as QtdeDocentesTerceir from professores where TP_TIPO_CONTRATACAO = 3 group by CO_ENTIDADE")

docentesCLT = psql.sqldf("select CO_ENTIDADE, count(TP_TIPO_CONTRATACAO) as QtdedocentesCLT from professores where TP_TIPO_CONTRATACAO = 4 group by CO_ENTIDADE")

docentesEfetivos.set_index('CO_ENTIDADE', inplace=True)

docentesTempo.set_index('CO_ENTIDADE', inplace=True)

docentesTerceir.set_index('CO_ENTIDADE', inplace=True)

docentesCLT.set_index('CO_ENTIDADE', inplace=True)
```

```python
variaveis_docentes = variaveis_docentes.join(docentesEfetivos)
variaveis_docentes = variaveis_docentes.join(docentesTempo)
variaveis_docentes = variaveis_docentes.join(docentesTerceir)
variaveis_docentes = variaveis_docentes.join(docentesCLT)
```
