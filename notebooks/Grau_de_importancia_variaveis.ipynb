```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset_completo = pd.read_csv('/content/drive/MyDrive/TCC/resultado_analise_merge_dataset_completo.csv')
dataset_se = pd.read_csv('/content/drive/MyDrive/TCC/resultado_analise_merge_dataset_completo_SE.csv')
dataset_al = pd.read_csv('/content/drive/MyDrive/TCC/resultado_analise_merge_dataset_completo_AL.csv')
dataset_ba = pd.read_csv('/content/drive/MyDrive/TCC/resultado_analise_merge_dataset_completo_BA.csv')

dataset_completo.rename(columns={"Unnamed: 0": "ID_ESCOLA"}, inplace=True)
dataset_se.rename(columns={"Unnamed: 0": "ID_ESCOLA"}, inplace=True)
dataset_al.rename(columns={"Unnamed: 0": "ID_ESCOLA"}, inplace=True)
dataset_ba.rename(columns={"Unnamed: 0": "ID_ESCOLA"}, inplace=True)

dataset_completo.set_index('ID_ESCOLA', inplace=True)
dataset_se.set_index('ID_ESCOLA', inplace=True)
dataset_ba.set_index('ID_ESCOLA', inplace=True)
dataset_al.set_index('ID_ESCOLA', inplace=True)
```

```python
def minmax(data):
    scaler = MinMaxScaler()
    data_scaler = scaler.fit_transform(data)
    return data_scaler
```

```python
def gerando_grafico_barra(X, features_importances):
  #data = dataset_se.drop(axis=1, columns=['Previsão Ideb'])
  plt.figure(figsize=(20, 100))
  importances = pd.Series(data=features_importances, index=X.columns)
  return sns.barplot(x=importances, y=importances.index, orient='h').set_title('Importância de cada feature', )
```

```python
def gerando_matriz_corr(dataset):
  plt.figure(figsize=(100, 100))
  sns.heatmap(dataset.corr(),
              annot = True,
              fmt = '.2f',
              cmap='Blues')
  plt.title('Correlação entre variáveis')
  return plt.show()
```

## Variáveis de Escolas Sergipanas

```python
X = dataset_se.drop(axis=1, columns=['Valor Real Ideb', 'Previsão Ideb'])
y = dataset_se['Previsão Ideb']

#Normalizando
X_transform = minmax(X)

#Treinando Modelo
random = RandomForestClassifier()
result = random.fit(X, y)

#Verificando grau de importância das variáveis
df_importance_al = pd.DataFrame(columns=X.columns, data=[result.feature_importances_])

df_importance_al.to_csv('importancia_variaveis_completo.csv')

#Gerando Gráficos
gerando_grafico_barra(X, result.feature_importances_)
```

## Variáveis de Escolas Alagoanas

```python
X = dataset_al.drop(axis=1, columns=['Valor Real Ideb', 'Previsão Ideb'])
y = dataset_al['Previsão Ideb']

#Normalizando
X_transform = minmax(X)

#Treinando Modelo
random = RandomForestClassifier()
result = random.fit(X, y)

#Verificando grau de importância das variáveis
df_importance_al = pd.DataFrame(columns=X.columns, data=[result.feature_importances_])

df_importance_al.to_csv('importancia_variaveis_al.csv')

#Gerando Gráficos
gerando_grafico_barra(X, result.feature_importances_)
```
