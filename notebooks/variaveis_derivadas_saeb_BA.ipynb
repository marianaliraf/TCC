```python
#Valor dos Parametros: base de dados, id da escola, pergunta do saeb e respostas
#Valor de Retorno: Variavel Derivada
#Funcao contabiliza as respostas para uma determinada pergunta
def criar_variavel_derivada(base, escola, pergunta, respostas):
    
    base_saeb = base[pergunta].value_counts().to_frame().T
    
    variavel_derivada = pd.DataFrame()
    
    for resposta in respostas:
        col =  pergunta+'_'+resposta
        if resposta in base_saeb.columns:
            variavel_derivada[col] = base_saeb[resposta]
        else:
            variavel_derivada[col] = 0
    
    variavel_derivada['ID_ESCOLA'] = escola

    variavel_derivada.set_index('ID_ESCOLA', inplace=True)
    '''
    base_saeb = base.groupby([pergunta]).count()
    aux = base_saeb[['ID_PROVA_BRASIL']]
    variavel_derivada = pd.DataFrame()
    
    for resposta in respostas:
        column = pergunta+'_'+ resposta
        variavel_derivada[column] = 0

    for resposta in aux.index:
        column = pergunta+'_'+ resposta
        var = aux.loc[aux.index == resposta]
        variavel_derivada[column] = var.values.reshape(-1)
    
    variavel_derivada['ID_ESCOLA'] = escola

    variavel_derivada.set_index('ID_ESCOLA', inplace=True)
    '''
    return variavel_derivada
```

```python
#Valor dos Parametros: base de dados, id da escola, perguntas do saeb e respostas
#Valor de Retorno: Variaveis Derivadas de uma escola
def variaveis_derivadas(base_dados, escola, perguntas, respostas):
    
    variavel = pd.DataFrame()
    
    for pergunta in perguntas: 
        if (variavel.empty):
            variavel = criar_variavel_derivada(base_dados, escola, pergunta, respostas)
        else:
            variavel = variavel.join(criar_variavel_derivada(base_dados, escola, pergunta, respostas))
    
    return variavel
```

```python
dataframe_duas_respostas = pd.DataFrame()
dataframe_tres_respostas = pd.DataFrame()
dataframe_quatro_respostas = pd.DataFrame()
dataframe_cinco_respostas = pd.DataFrame()
dataframe_seis_respostas = pd.DataFrame()
dataframe_sete_respostas = pd.DataFrame()
dataframe_oito_respostas = pd.DataFrame()
dataframe_doze_respostas = pd.DataFrame()
```

```python
#gerando variaveis derivadas

for escola in escolas:
    base = base_saeb_ba.loc[base_saeb_ba['ID_ESCOLA'] == escola]
    
    dataframe_duas_respostas = dataframe_duas_respostas.append(variaveis_derivadas(base, escola, perguntas_duas_respostas, respostas[:2]))
    
    dataframe_tres_respostas = dataframe_tres_respostas.append(variaveis_derivadas(base, escola, perguntas_tres_respostas, respostas[:3]))

    dataframe_quatro_respostas = dataframe_quatro_respostas.append(variaveis_derivadas(base, escola, perguntas_quatro_respostas, respostas[:4]))

    dataframe_cinco_respostas = dataframe_cinco_respostas.append(variaveis_derivadas(base, escola, perguntas_cinco_respostas, respostas[:5]))
    
    dataframe_seis_respostas = dataframe_seis_respostas.append(variaveis_derivadas(base, escola, perguntas_seis_respostas, respostas[:6]))

    dataframe_sete_respostas = dataframe_sete_respostas.append(variaveis_derivadas(base, escola, perguntas_sete_respostas, respostas[:7]))

    dataframe_oito_respostas = dataframe_oito_respostas.append(variaveis_derivadas(base, escola, perguntas_oito_respostas, respostas[:8]))

    dataframe_doze_respostas = dataframe_doze_respostas.append(variaveis_derivadas(base, escola, perguntas_doze_respostas, respostas[:12]))
```

```python
#Função que verifica se o ID de uma escola é inválido
#Segundo o Dicionário de Dados da base o SAEB 6 - Os códigos dos Municípios e 
#das Escolas que começam com o dígito "6" são máscaras, isto é, são códigos fictícios.
def retira_id_ficticio(id_escola):
    if id_escola[0] == "6":
        return True
    else:
        return False
```

```python
#Se retirou as escolas fictícias
for escola in dataframe_var_derivadas.index:
    if (retira_id_ficticio(str(escola))):
        dataframe_var_derivadas.drop(escola, inplace=True)
        

dataframe_var_derivadas
```
